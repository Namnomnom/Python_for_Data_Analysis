## Handling JSON

```python
# a simple employee for json 3 employees
employee_data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [30, 25, 35],
    "department": ["HR", "Engineering", "Marketing"],
    "salary": [70000, 80000, 90000]
}
df_employee = pd.DataFrame(employee_data)
df_employee.head()
```

```python
df_employee.to_json("sample_employees.json", orient="records", lines=True) # speichert eine DataFrame als JSON-Datei
```

```python
#use the read_json function to read the JSON file
df_employees = pd.read_json("sample_employees.json", orient="records", lines=True)
df_employees.head()  # Display the first two rows of the DataFrame
```

```python
# Display the shape of the DataFrame    
df_employees.shape  # Display the shape of the DataFrame
# Display the data types of each column in the DataFrame
#df_employees.dtypes
```

```python

#not a best practice to use

def format_salary(s):
    return f"${s:,}"
def format_age(a):
    return f"{a} years old"

df_employees.head().assign(
    salary=lambda x: x['salary'].apply(format_salary),  # Format salary with commas / goes through each value in the 'salary' column and applies the format_salary function
    age=lambda x: x['age'].apply(format_age)   
).head()
```

```python
# assignment -user defined function

```

```python
#use for loop to iterate over rows and print formatted strings
for index, row in df_employees.iterrows():
    print(f"{row['name']} works in {row['department']} and earns {row['salary']:,} annually.")  
```

```python
#use apply with a lambda function to format the salary
df_employees['salary_formatted'] = df_employees['salary'].apply(lambda x: f"${x:,.2f}") 
df_employees.head()  # Display the DataFrame with the formatted salary column
```

```python
#use for loop to iterate over rows and print formatted strings
for index, row in df_employees.iterrows():
    print(f"{row['name']} works in {row['department']} and earns {row['salary_formatted']} annually.")  
```

```python

# Save the modified DataFrame back to JSON
# what is records and lines in json
# 'records' format creates a list of dictionaries, where each dictionary represents a row in the DataFrame.
# 'lines=True' writes each record on a new line, making it easier to read and process large datasets line by line.
df_employees.to_json("sample_employees_modified.json", orient="records", lines=True)
```

## Reading from SQLite

```python

import pandas as pd
import sqlite3

conn = sqlite3.connect("sample_school.db")
```

```python
#add data to the database
# Create a sample DataFrame for students    
students_data = {
    "student_id": [1, 2, 3, 4],
    "name": ["Alice", "Bob", "Charlie", "David"],
    "age": [20, 21, 22, 23],
    "major": ["Computer Science", "Mathematics", "Physics", "Chemistry"]
}
df_students = pd.DataFrame(students_data)
df_students.head()  # Display of the DataFrame
```

```python
# Save the DataFrame to a SQLite database
df_students.to_sql("students", conn, if_exists="replace", index=False)
```

```python
# Query the database to retrieve data from the students table
df_students_query = pd.read_sql_query("SELECT * FROM students", conn)
df_students_query.head()  # Display the first two rows of the queried DataFrame
```

```python
#create a sample DataFrame for courses
courses_data = {
    "course_id": [1, 2, 3, 4],
    "course_name": ["Data Structures", "Calculus", "Quantum Mechanics", "Organic Chemistry"],
    "credits": [3, 4, 3, 4]
}
df_courses = pd.DataFrame(courses_data) 
# Save the DataFrame to the SQLite database
df_courses.to_sql("courses", conn, if_exists="replace", index=False)
df_courses.head()  # Display the DataFrame
```

```python
#merge the students and courses DataFrames
df_merged = pd.merge(df_students, df_courses, left_on="student_id", right_on="course_id", how="inner")
df_merged.head()  # Display the first two rows of the merged DataFrame
```

```python
#make a query to get students with age greater than 21
df_students_age_query = pd.read_sql_query("SELECT * FROM students WHERE age > 21", conn)
df_students_age_query.head()  # Display the first two rows of the queried DataFrame
```

```python
#add score to the students DataFrame
df_students['score'] = [85, 90, 78, 88]  # Add a new column 'score' to the students DataFrame
# Save the updated DataFrame back to the SQLite database
df_students.to_sql("students", conn, if_exists="replace", index=False)
```

```python
df_students.head()
```

```python
# add status column based on score
df_students_age_query['status'] = df_students_age_query['score'].apply(lambda x: 'Pass' if x >= 80 else 'Fail')
df_students_age_query.head()  # Display the first two rows of the updated DataFrame
```

```python
#merge students_age_query to students
df_students_age_query_merged = pd.merge(df_students_age_query, df_courses, left_on="student_id", right_on="course_id", how="inner")
df_students_age_query_merged.head()  # Display the first two rows of the merged DataFrame 
```

```python
#write the merged DataFrame to a new table in the SQLite database
df_students_age_query_merged.to_sql("students_courses", conn, if_exists="replace", index=False)
```

```python
# Query the new table to verify the data
df_verified = pd.read_sql("SELECT * FROM students_courses", conn)
df_verified.head()  # Display the first two rows of the verified DataFrame
```

```python
# Close the database connection
conn.close()  # Close the SQLite database connection
```
